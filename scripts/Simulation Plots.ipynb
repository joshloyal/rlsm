## Sender and Reciever Parameters (Figure 2)

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 6))

axs[0].plot(x, se_mse, marker='o', linestyle='-', color='black')
axs[0].fill_between(x, se1, se2, color='grey', alpha=0.5, label='Shadow')
axs[0].set_title(r'Sender Parameters ($s_i$)', fontsize=20)
axs[0].set_xlabel('Number of Nodes (n)', fontsize=20)
axs[0].set_ylabel('MSE', fontsize=20)
axs[0].grid(False)
axs[0].tick_params(labelsize=20)

axs[1].plot(x, re_mse, marker='o', linestyle='-', color='black')
axs[1].fill_between(x, re1, re2, color='grey', alpha=0.5, label='Shadow')
axs[1].set_title(r'Receiver Parameters ($r_i$)', fontsize=20)
axs[1].set_xlabel('Number of Nodes (n)', fontsize=20)
axs[1].grid(False)
axs[1].tick_params(labelsize=20)
```

## Reciprocity Parameters (Figure 3)

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 6))

axs[0].plot(x, recip_coef_mse, marker='o', linestyle='-', color='black')
axs[0].fill_between(x, rc1, rc2, color='grey', alpha=0.5, label='Shadow')
axs[0].set_title(r'Baseline Reciprocity ($\rho$)', fontsize=20)
axs[0].set_xlabel('Number of Nodes (n)', fontsize=20)
axs[0].set_ylabel(r'$|\rho - \hat{\rho}|$', fontsize=20)
axs[0].grid(False)
axs[0].tick_params(labelsize=20)

axs[1].plot(x, dist_coef_mse, marker='o', linestyle='-', color='black')
axs[1].fill_between(x, dc1, dc2, color='grey', alpha=0.5, label='Shadow')
axs[1].set_title(r'Distance Coefficient ($\phi$)', fontsize=20)
axs[1].set_xlabel('Number of Nodes (n)', fontsize=20)
axs[1].set_ylabel(r'$|\phi - \hat{\phi}|$', fontsize=20)
axs[1].grid(False)
axs[1].tick_params(labelsize=20)
```
